```python
#import library
import numpy as np
import pandas as pd
```

```python
#import dataset
url = 'https://raw.githubusercontent.com/ShvrCol/DS_dataset_diabates/main/full_data.csv'
df = pd.read_csv(url, parse_dates=['date'])
df.head()
```

```python
#replace NaN values with 0
df_new = df.replace(np.nan, 0)
```

```python
#convert float to int
df_new['new_cases'] = df_new['new_cases'].astype(int)
df_new['new_deaths'] = df_new['new_deaths'].astype(int)
df_new['total_deaths'] = df_new['total_deaths'].astype(int)
```

```python
df_new
```

```python
#Create a pandas groupby DataFrame
countrytots = df_new.groupby(['location'])
type(countrytots)
```

```python
#Create DataFrames for the first and last rows of each country
countrytots.first().iloc[0:10, 0:10]
```

```python
countrytots.last().iloc[0:10, 0:10]
```

```python
type(countrytots.last())
```

```python
#Get all the rows for a country
countrytots.get_group('Spain').iloc[0:10, 0:10]
```

```python
#Loop through the groups
for name, group in countrytots:
    if (name in ['Malta','Kuwait']):
        print(group.iloc[0:10, 0:6])
```

```python
#Show the number of rows for each country
countrytots.size()
```
